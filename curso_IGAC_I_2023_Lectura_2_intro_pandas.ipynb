## ***Número predial nacional.***

El **número predial nacional (npn)** es un código que permite identificar cada uno de los predios o bienes inmuebles existentes en el territorio colombiano sin posibilidad de equivocación en virtud de su condición geográfica, sus condiciones y atributos.

**Nota:** El IGAC adopta el Número Predial Nacional a partir de enero de 2014.

### ***Componentes.***

El número predial está conformado por ocho componentes

1. Código del departamento.

2. Código del municipio.

3. Zona.

4. Sector, Comuna, Barrio.

5. Manzana o vereda.

6. Terreno.

7. Condicion del predio.

8. Número de construcción/PH.

Estas ocho posiciones agrupan un total de 30 posiciones que se desglosan asi:

### ***Intro funciones.***

En Python, una función es un bloque de código que realiza una tarea específica y que puede ser llamado desde otras partes del programa. Las funciones son una forma de organizar y reutilizar el código en programas más grandes y complejos.

La sintaxis básica de una función en Python es la siguiente:

```python
def mi_primera_funcion(num):
    """
    Eleva un número a su cuadrado.
    Input: num:(número) ....
    Output: El número al cuadrado
    """
    square = num*num
    return square

mi_primera_funcion(8)
```

Aquí, `mi_primera_funcion` es el nombre de la función, que se utiliza para llamarla desde otras partes del programa. `num` argumentos son los valores que se pasan a la función como entrada. El bloque de código que se encuentra en el cuerpo de la función es el que realiza la tarea específica. El resultado de la función se devuelve mediante la instrucción return.

```python
## En Python, una función lambda es una función anónima y sin nombre que se utiliza para crear funciones simples 
## y rápidas sin la necesidad de definir una función completa. Las funciones lambda se crean utilizando la palabra 
## clave lambda, seguida de los parámetros de entrada y la expresión que se evalúa para devolver el resultado.
## La sintaxis básica de una función lambda en Python es la siguiente:
andres = lambda x: x**2
andres(5)
```

```python
## Vamos a construir estas nuevas columnas
df_registro1['DIVIPOLA'] = df_registro1['NPN'].apply(lambda x: str(x)[0:5])
df_registro1['ZONA'] = df_registro1['NPN'].apply(lambda x: str(x)[5:7])
df_registro1['SECTOR_COMUNA_BARRIO'] = df_registro1['NPN'].apply(lambda x: str(x)[7:13])
df_registro1['MANZANA_VEREDA'] = df_registro1['NPN'].apply(lambda x: str(x)[13:17])
df_registro1['TERRENO'] = df_registro1['NPN'].apply(lambda x: str(x)[17:21])
df_registro1['CONDICION_PREDIO'] = df_registro1['NPN'].apply(lambda x: str(x)[21])
df_registro1['NUMERO_CONSTRUCCION'] = df_registro1['NPN'].apply(lambda x: str(x)[22:])
### Columns to move
columns_move = ['DIVIPOLA', 'ZONA', 'SECTOR_COMUNA_BARRIO', 'MANZANA_VEREDA', 'TERRENO', \
                'CONDICION_PREDIO', 'NUMERO_CONSTRUCCION']
columns = [df_registro1.pop(col) for col in columns_move]
for idx, col in enumerate(columns):
    df_registro1.insert(idx+1, col.name, col)
display(df_registro1)
```
