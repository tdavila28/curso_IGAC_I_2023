```python
## Vamos a construir un diccionario y una función para realizar unos descriptivos más detallados.
mapper_condition = {'0':'NoRegPH', '2':'PredioInformal', '3':'UsoPubNoVias', '4':'Vias', '5':'MejoraPredioAjeno',
                    '6':'MejAjenoPH', '7':'ParqCemente', '8':'PH_Condominio', '9':'PH'}
## Vamos a definir tres zonas: Rural --->"00", Urbano ---> "01", Corregimientos
def zone_classification(zone):
    '''
    Esta función reasigna la zona en tres grandes grupos: rural, urbano y corregimientos
    Input: str (00, 01, 02, ..., 09)  <--- SE PRODRÍA PENSAR EN UN mapper como el caso anterior.
    Output: str con los códigos en palabras.
    '''
    if zone == '00':
        return 'rural'
    elif zone == '01':
        return 'urbano'
    else:
        return 'corregimiento'
##
df_plot = df_r1.copy()
df_plot['CONDICION_PREDIO'] = df_plot['CONDICION_PREDIO'].map(mapper_condition)
df_plot['ZONA'] = df_plot['ZONA'].apply(lambda x: zone_classification(x))
df_plot
```

```python
##
df_plot['ZONA'].value_counts()
```

```python
##
df_plot['CONDICION_PREDIO'].value_counts(normalize=1)
```

```python
## Agrupaciones
df_r1.groupby(by=['TOTAL_REGISTROS'],).size().to_frame('cantidad')
```

```python
## Veamos cuantos y que % de los predios tiene uno más registros.
temp = df_r1.groupby(by=['TOTAL_REGISTROS']).agg(CANTIDAD=('NPN', 'count'))
temp['PORCENTAJE'] = round(temp['CANTIDAD']/temp['CANTIDAD'].sum()*100, 3)
temp.loc['Totales'] = temp.sum()
## Mejorando la salida. Comente la linea 6 y 7 y observe la diferencia.
temp['PORCENTAJE'] = temp['PORCENTAJE'].astype('str') + ' %'
temp = temp.astype({'CANTIDAD':'int'})
display(temp)
```

```python
## Escribimos el pd.DataFrame en formato Excel con las nuevas "mejoras"
df_plot.to_excel('data/nuqui_r1_plot.xlsx', index=False, sheet_name='NuquiR1')
```
